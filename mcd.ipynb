```python
df = pd.read_csv('menu2.csv')
```

```python
data = df.drop(['Item','Serving_Size'],axis=1)
data.info()
```

```python
data['Sodium(%DailyValue)'] = data['Sodium(%DailyValue)'].fillna(data['Sodium(%DailyValue)'].mean())
data['Calories'] = data['Calories'].fillna(data['Calories'].mean())
data['Calcium(%DailyValue)'] = data['Calcium(%DailyValue)'].fillna(data['Calcium(%DailyValue)'].mean())
data['Vitamin_C(%DailyValue)'] = data['Vitamin_C(%DailyValue)'].fillna(data['Vitamin_C(%DailyValue)'].mean())
data['Vitamin_A(%DailyValue)'] = data['Vitamin_A(%DailyValue)'].fillna(data['Vitamin_A(%DailyValue)'].mean())
data['Protein'] = data['Protein'].fillna(data['Protein'].mean())
data['Sugars'] = data['Sugars'].fillna(data['Sugars'].mean())
data['Dietary_Fiber(%DailyValue)'] = data['Dietary_Fiber(%DailyValue)'].fillna(data['Dietary_Fiber(%DailyValue)'].mean())
data['Dietary_Fiber'] = data['Dietary_Fiber'].fillna(data['Dietary_Fiber'].mean())
data['Carbohydrates(%DailyValue)'] = data['Carbohydrates(%DailyValue)'].fillna(data['Carbohydrates(%DailyValue)'].mean())
data['Dietary_Fiber'] = data['Dietary_Fiber'].fillna(data['Dietary_Fiber'].mean())
data['Carbohydrates(%DailyValue)'] = data['Carbohydrates(%DailyValue)'].fillna(data['Carbohydrates(%DailyValue)'].mean())
data['Carbohydrates'] = data['Carbohydrates'].fillna(data['Carbohydrates'].mean())
data['Iron(%DailyValue)'] = data['Iron(%DailyValue)'].fillna(data['Iron(%DailyValue)'].mean())
data['Sodium'] = data['Sodium'].fillna(data['Sodium'].mean())
data['Cholesterol(%DailyValue)'] = data['Cholesterol(%DailyValue)'].fillna(data['Cholesterol(%DailyValue)'].mean())
data['Cholesterol'] = data['Cholesterol'].fillna(data['Cholesterol'].mean())
data['Trans_Fat'] = data['Trans_Fat'].fillna(data['Trans_Fat'].mean())
data['Saturated_Fat(%DailyValue)'] = data['Saturated_Fat(%DailyValue)'].fillna(data['Saturated_Fat(%DailyValue)'].mean())
data['Saturated_Fat'] = data['Saturated_Fat'].fillna(data['Saturated_Fat'].mean())
data['Total_Fat(%DailyValue)'] = data['Total_Fat(%DailyValue)'].fillna(data['Total_Fat(%DailyValue)'].mean())
data['Total_Fat'] = data['Total_Fat'].fillna(data['Total_Fat'].mean())
data['Calories_from_Fat'] = data['Calories_from_Fat'].fillna(data['Calories_from_Fat'].mean())
```

```python
data['Category'] = data['Category'].map({'Coffee & Tea':1, 'Breakfast':2, 'Smoothies & Shakes':3, 'Beverages':4, 'Chicken & Fish':5, 'Beef & Pork':6, 'Snacks & Sides':7, 'Desserts':8, 'Salads':9})
```

```python
X = data.drop (columns='Category', axis=1)
Y = data['Category']
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
std = scaler.transform(X)
```

```python
X = std
Y = data['Category']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, stratify=Y, random_state=2)
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
X_train_pred = model.predict(X_train)
```

```python
training_data_accuracy = accuracy_score(X_train_pred, Y_train)
```

```python
print('akurasi : ', training_data_accuracy)
```

```python
X_test_pred = model.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(X_test_pred, Y_test)
```

```python
print('akurasi : ', test_data_accuracy)
```
